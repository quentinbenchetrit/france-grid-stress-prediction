```python
import zipfile
import pandas as pd

zip_path = "/home/onyxia/france-grid-stress-prediction/data/raw*/consommation/Historique_consommation_INST_2016.zip"

with zipfile.ZipFile(zip_path, "r") as z:
    inner = z.namelist()[0]   
    with z.open(inner) as f:
        df_raw = pd.read_excel(f)   

print(df_raw.shape)
print(df_raw.head())
print(df_raw.tail())
```

```python
df = df_raw.copy()

# 1) On renomme les colonnes
df.columns = [f"col_{i}" for i in range(df.shape[1])]

# 2) On garde les lignes où col_0 est une vraie date (format jour/mois/année)
dates = pd.to_datetime(df["col_0"], format="%d/%m/%Y", errors="coerce")
df = df[dates.notna()].copy()
df["date"] = dates

# 3) On renomme les deux premières colonnes utiles
df = df.rename(columns={
    "col_0": "date_str",
    "col_1": "statut"
})

# 4) On enlève les colonnes entièrement vides (si besoin)
df = df.dropna(axis=1, how="all")

print(df.shape)
print(df[["date", "statut"]].head())
```

```python
import numpy as np

# Colonnes de pas de temps = toutes sauf 'date', 'date_str', 'statut'
time_cols = [c for c in df.columns if c not in ["date", "date_str", "statut"]]

# On passe en long : une ligne = (date, slot_index, valeur)
df_long = df.melt(
    id_vars=["date", "statut"],
    value_vars=time_cols,
    var_name="slot_col",
    value_name="load_mw"
)

# On ordonne par date + colonne
df_long = df_long.sort_values(["date", "slot_col"]).reset_index(drop=True)

# Index de pas de temps par jour : 0,1,2,... pour chaque date
df_long["slot_index"] = df_long.groupby("date").cumcount()

# Hypothèse : chaque slot = 30 minutes (à adapter si c'est 1h)
df_long["datetime"] = df_long["date"] + pd.to_timedelta(df_long["slot_index"] * 30, unit="m")

# On garde l’essentiel
df_long = df_long[["datetime", "date", "statut", "slot_index", "load_mw"]]

print(df_long.head())
print(df_long.tail())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import glob

# --- Description Globale ---
# Ce script charge les données de consommation, convertit la colonne temporelle
# au format datetime, puis trace la courbe complète de la charge électrique sur l'année 2010.
# L'objectif est de visualiser la saisonnalité annuelle et les pics de consommation.

# 1. Chargement des données avec gestion du chemin générique (*)
file_pattern = '/home/onyxia/work/france-grid-stress-prediction/data/raw*/consommation/consommation_2010_long.csv'
file_paths = glob.glob(file_pattern)
if not file_paths:
    raise FileNotFoundError("Aucun fichier trouvé correspondant au chemin spécifié.")
    
df = pd.read_csv(file_paths[0])

# 2. Prétraitement : Conversion de la colonne date
df['datetime'] = pd.to_datetime(df['datetime'])

# 3. Visualisation : Courbe de charge chronologique
plt.figure(figsize=(15, 6))
plt.plot(df['datetime'], df['load_mw'], linewidth=0.5, color='#1f77b4')
plt.title('Tendance globale de la consommation électrique (2010)')
plt.ylabel('Consommation (MW)')
plt.xlabel('Date')
plt.grid(True, alpha=0.3)

# Sauvegarde et affichage
plt.tight_layout()
plt.savefig('tendance_globale.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob

# --- Description Globale ---
# Ce script examine la saisonnalité mensuelle. Il extrait le mois de chaque observation
# et utilise des boxplots pour afficher la distribution de la consommation pour chaque mois.
# Cela permet de comparer la variabilité et le niveau médian entre l'hiver et l'été.

# 1. Chargement des données
file_pattern = '/home/onyxia/work/france-grid-stress-prediction/data/raw*/consommation/consommation_2010_long.csv'
df = pd.read_csv(glob.glob(file_pattern)[0])
df['datetime'] = pd.to_datetime(df['datetime'])

# 2. Feature Engineering : Extraction du mois
df['month'] = df['datetime'].dt.month

# 3. Visualisation : Boxplots mensuels
plt.figure(figsize=(12, 6))
# Palette 'coolwarm' pour symboliser le froid (hiver) et le chaud (été)
sns.boxplot(x='month', y='load_mw', data=df, palette="coolwarm")

plt.title('Distribution mensuelle de la consommation (Effet saisonnier)')
plt.ylabel('Consommation (MW)')
plt.xlabel('Mois')
plt.grid(True, axis='y', alpha=0.3)

# Sauvegarde et affichage
plt.tight_layout()
plt.savefig('saisonnalite_mensuelle.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob

# --- Description Globale ---
# Ce script se concentre sur l'échelle hebdomadaire et journalière.
# Il calcule la consommation moyenne pour chaque créneau horaire (ex: 12h00, 12h30...)
# séparément pour chaque jour de la semaine. Les courbes résultantes permettent
# d'identifier les profils types (pointe du matin, pointe du soir, baisse le week-end).

# 1. Chargement des données
file_pattern = '/home/onyxia/work/france-grid-stress-prediction/data/raw*/consommation/consommation_2010_long.csv'
df = pd.read_csv(glob.glob(file_pattern)[0])
df['datetime'] = pd.to_datetime(df['datetime'])

# 2. Feature Engineering : Jour de la semaine et Heure numérique
# Ordonner les jours de la semaine
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['day_name'] = pd.Categorical(df['datetime'].dt.day_name(), categories=days_order, ordered=True)

# Création d'une heure décimale pour l'axe X (ex: 8h30 devient 8.5)
df['hour_float'] = df['datetime'].dt.hour + df['datetime'].dt.minute / 60

# 3. Agrégation : Moyenne par jour et par heure
daily_profile = df.groupby(['day_name', 'hour_float'])['load_mw'].mean().reset_index()

# 4. Visualisation : Courbes superposées
plt.figure(figsize=(15, 8))
sns.lineplot(data=daily_profile, x='hour_float', y='load_mw', hue='day_name', palette='tab10', linewidth=2)

plt.title('Profils de consommation moyens par jour de la semaine')
plt.ylabel('Consommation Moyenne (MW)')
plt.xlabel('Heure de la journée')
plt.xticks(range(0, 25, 2)) # Graduation toutes les 2h
plt.legend(title='Jour', loc='upper left')
plt.grid(True, alpha=0.3)

# Sauvegarde et affichage
plt.tight_layout()
plt.savefig('profils_hebdomadaires.png')
plt.show()
```
