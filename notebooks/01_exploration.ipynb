```python
import zipfile
import pandas as pd

zip_path = "/home/onyxia/france-grid-stress-prediction/data/raw*/consommation/Historique_consommation_INST_2016.zip"

with zipfile.ZipFile(zip_path, "r") as z:
    inner = z.namelist()[0]   
    with z.open(inner) as f:
        df_raw = pd.read_excel(f)   

print(df_raw.shape)
print(df_raw.head())
print(df_raw.tail())
```

```python
df = df_raw.copy()

# 1) On renomme les colonnes
df.columns = [f"col_{i}" for i in range(df.shape[1])]

# 2) On garde les lignes où col_0 est une vraie date (format jour/mois/année)
dates = pd.to_datetime(df["col_0"], format="%d/%m/%Y", errors="coerce")
df = df[dates.notna()].copy()
df["date"] = dates

# 3) On renomme les deux premières colonnes utiles
df = df.rename(columns={
    "col_0": "date_str",
    "col_1": "statut"
})

# 4) On enlève les colonnes entièrement vides (si besoin)
df = df.dropna(axis=1, how="all")

print(df.shape)
print(df[["date", "statut"]].head())
```

```python
import numpy as np

# Colonnes de pas de temps = toutes sauf 'date', 'date_str', 'statut'
time_cols = [c for c in df.columns if c not in ["date", "date_str", "statut"]]

# On passe en long : une ligne = (date, slot_index, valeur)
df_long = df.melt(
    id_vars=["date", "statut"],
    value_vars=time_cols,
    var_name="slot_col",
    value_name="load_mw"
)

# On ordonne par date + colonne
df_long = df_long.sort_values(["date", "slot_col"]).reset_index(drop=True)

# Index de pas de temps par jour : 0,1,2,... pour chaque date
df_long["slot_index"] = df_long.groupby("date").cumcount()

# Hypothèse : chaque slot = 30 minutes (à adapter si c'est 1h)
df_long["datetime"] = df_long["date"] + pd.to_timedelta(df_long["slot_index"] * 30, unit="m")

# On garde l’essentiel
df_long = df_long[["datetime", "date", "statut", "slot_index", "load_mw"]]

print(df_long.head())
print(df_long.tail())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# --- Description Globale ---
# Ce script examine la saisonnalité mensuelle. Il extrait le mois de chaque observation
# et utilise des boxplots pour afficher la distribution de la consommation pour chaque mois.
# Cela permet de comparer la variabilité et le niveau médian entre l'hiver et l'été.

PROJECT_ROOT = Path("/home/onyxia/france-grid-stress-prediction")
DATA_PROCESSED = PROJECT_ROOT / "data" / "processed"

# 1. Chargement des données (version nettoyée)
df = pd.read_parquet(DATA_PROCESSED / "consommation_clean.parquet")
df["datetime"] = pd.to_datetime(df["datetime"])
df = df.sort_values("datetime")

# 2. Filtre sur 2010 (équivalent à consommation_2010_long.csv, mais propre)
df_2010 = df[df["datetime"].dt.year == 2010].copy()

# 3. Feature Engineering : Extraction du mois
df_2010["month"] = df_2010["datetime"].dt.month

# 4. Visualisation : Boxplots mensuels
plt.figure(figsize=(12, 6))
sns.boxplot(x="month", y="load_mw", data=df_2010, palette="coolwarm")

plt.title("Distribution mensuelle de la consommation (Effet saisonnier) – 2010")
plt.ylabel("Consommation (MW)")
plt.xlabel("Mois")
plt.grid(True, axis="y", alpha=0.3)

# Sauvegarde et affichage
plt.tight_layout()
plt.savefig("saisonnalite_mensuelle.png")
plt.show()
```

```python
# Ordre académique des jours
days_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

df_2010["day_name"] = pd.Categorical(
    df_2010["datetime"].dt.day_name(),
    categories=days_order,
    ordered=True
)

# Heure décimale
df_2010["hour_float"] = (
    df_2010["datetime"].dt.hour +
    df_2010["datetime"].dt.minute / 60
)

# Agrégation
daily_profile = (
    df_2010
    .groupby(["day_name", "hour_float"])["load_mw"]
    .mean()
    .reset_index()
)

# Plot
plt.figure(figsize=(15, 8))
sns.lineplot(
    data=daily_profile,
    x="hour_float",
    y="load_mw",
    hue="day_name",
    palette="tab10",
    linewidth=2
)

plt.title("Profils moyens de consommation par jour de la semaine (2010)")
plt.xlabel("Heure de la journée")
plt.ylabel("Consommation moyenne (MW)")
plt.xticks(range(0, 25, 2))
plt.grid(True, alpha=0.3)
plt.legend(title="Jour", loc="upper left")

plt.tight_layout()
plt.savefig("profils_hebdomadaires.png")
plt.show()
```
